#### Questão 2: O código abaixo simula o lançamento de um dado 10000 vezes. Os resultados dos lançamentos são armazenados em uma lista 'lancamentos'. Complete o código dentro da área delimitada, fazendo a verificação de quais números são pares, fazendo a contagem de pares e armazenando o valor na variável 'conta_pares'. Calcule a proporção de números pares em relação ao total de lançamentos feitos. O resultado é próximo ao obtido na questão 1?

```python
import random

lancamentos = []
conta_pares = 0
proporcao = 0

for i in range(10000):
    lancamentos.append(random.randint(1,6))
    
for lancamento in lancamentos:
    ### COMEÇO DO CÓDIGO ### 
    if lancamento%2==0:
        conta_pares += 1
    proporcao = conta_pares/10000
    ### FIM DO CÓDIGO ###
print('A proporçao é igual a ', proporcao)
```

#### Questão 3: Complete o código abaixo, que calcula e imprime o número de permutações (a ordem importa!) possíveis com as letras A,B,C,D.

```python
#https://www.acervolima.com.br/2020/12/permutacao-e-combinacao-em-python.html
import math

### COMEÇO DO CÓDIGO ### 
from itertools import permutations

permutacoes = permutations(["A","B","C","D"])
soma = 0

for i in list(permutacoes):
    soma = soma + 1
    #print (i)
    
permutacoes = soma
### FIM DO CÓDIGO ###

print('O número de permutações possíveis é:', permutacoes)
```

#### Questão 12: Escreva um trecho de código para calcular o z-score da média amostral:

```python
import math   

### COMEÇO DO CÓDIGO ### 
x = 2.875
mean = 3.125
std = 0.700
n = 40
z_score = (x - mean)/(std/math.sqrt(n))

### FIM DO CÓDIGO ###
print('O z-score da média amostral é de:', z_score)
```
