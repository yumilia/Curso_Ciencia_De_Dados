### Aula 25 - Exemplo simples de Decision Tree

```python
from sklearn import tree
```

```python
# features = [[140, 'smooth'], [130, 'smooth'], [170, 'bumpy'], [150, 'bumpy']]
# labels = ['apple', 'apple', 'orange', 'orange']

# Para as features, encoding de 1  para 'smooth' e 0 para 'bumpy'
# Para labels, encoding de 0 para 'apple' e 1 para 'orange'
features = [[140, 1], [130, 1], [170, 0], [150, 0]]
labels = [0, 0, 1, 1]
```

```python
# Criando o classificador
clf = tree.DecisionTreeClassifier()
```

```python
# Treinando o modelo
clf = clf.fit(features, labels)
```

```python
# Fazendo predição
print(clf.predict([[1000, 1]]))
```

```python
'''Ao plotarmos a árvore, verificamos que somente a coluna X[1] é utilizada pelo modelo, 
visto que ela já faz uma separação perfeita dos dados de treinamento. 
'''
tree.plot_tree(clf)
```
