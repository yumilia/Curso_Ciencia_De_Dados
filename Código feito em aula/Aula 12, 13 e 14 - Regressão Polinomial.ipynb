# Aula 12 e 13 - Regressão Polinominal

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Criando dataset baseado em uma parábola
#Adiciona ruído com distribuição normal
x = 10*np.random.normal(0,1,70)
y = 10*(-x**2) + np.random.normal(-100,100,70)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split (x, y, test_size = 0.2)
```

```python
#Gráfico de dispersão do dataset
plt.figure(figsize=(10,5))
plt.scatter(x_train, y_train, color='r', label='Train')
plt.scatter(x_test, y_test, color='b', label='Test')
plt.xlabel('x_train')
plt.ylabel('y_train')
plt.legend()
plt.title('Dispersão dos conjuntos de treinamentos e de teste')
plt.show()
```

## Regressão Linear

```python
from sklearn.linear_model import LinearRegression

x_train = x_train.reshape(-1, 1)
y_train = y_train.reshape(-1, 1)

x_test = x_test.reshape(-1, 1)
y_test = y_test.reshape(-1, 1)

lm = LinearRegression()
lm.fit(x_train, y_train)
```

```python
#Resultado da Regressão Linear

y_train_pred = lm.predict(x_train)
```

```python
#Gráfico da Regressão Linear no conjunto de treinamento
plt.figure(figsize=(10,5))
plt.scatter(x_train, y_train, color = 'r', label='Train')
plt.plot(x_train, y_train_pred, color = 'g', label='Prediction')
plt.xlabel('x_train')
plt.ylabel('y_train')
plt.legend()
plt.title('Regressão Linear no conjunto de treinamento')
plt.show()
```

```python
#Cálculo do R**2
r2 = lm.score(x_train**2, y_train)
print("R quadrado para Regressão Linear simples: ",r2)
```

## Regressão Polinomial

```python
#Ordenar o dataset
trainset = np.concatenate((x_train,y_train),axis=1)
trainset = trainset[trainset[:,0].argsort()]

testset = np.concatenate((x_test,y_test),axis=1)
testset = testset[testset[:,0].argsort()]

x_train = trainset[:,0].reshape(-1,1)
y_train = trainset[:,1].reshape(-1,1)
x_test = testset[:,0].reshape(-1,1)
y_test = testset[:,1].reshape(-1,1)
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly_reg = PolynomialFeatures(degree=2)
x_train_poly = poly_reg.fit_transform(x_train)
poly_reg.fit(x_train_poly, y_train)

lm_poly = LinearRegression()
lm_poly.fit(x_train_poly, y_train)
```

```python
y_train_poly_pred = lm_poly.predict(x_train_poly)
```

```python
#Gráfico da regressão polinomial no conjunto de treinamento
plt.figure(figsize=(10,5))
plt.scatter(x_train, y_train, color = 'r', label='Train set')
plt.plot(x_train, y_train_poly_pred, color = 'g', label='Prediction')
plt.xlabel('x_train')
plt.ylabel('y_train')
plt.legend()
plt.title('Regressão Polinomial no conjunto de treinamento ')
plt.show()
```

```python
#Testando o algoritmo no conjunto de teste
# modelo.fit(x_treinamento, y_treinamento)
# y_predicao = modelo.predict(x_test)
# comparar y_predicao com y_teste

x_test_poly = poly_reg.fit_transform (x_test)
y_test_poly_pred = lm_poly.predict(x_test_poly)
```

```python
#Gráfico da regressão polinomial no conjunto de teste
plt.figure(figsize=(10,5))
plt.scatter(x_test, y_test, color = 'r', label='Test set')
plt.plot(x_test, y_test_poly_pred, color = 'g', label='Prediction')
plt.xlabel('x_test')
plt.ylabel('y_test')
plt.legend()
plt.title('Regressão Polinomial no conjunto de treinamento ')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
rmse  = np.sqrt(mean_squared_error(y_test, y_test_poly_pred))
print("Root mean squared error para a Regressão Polimiomial, no conjunto de teste:", rmse)
```

### Testando outros valores de grau (degree) de polinômio

```python
rmses = []
degrees = np.arange(1, 5)
min_rmse = 1e10
min_deg = 0
```

```python
for deg in degrees:
    #Treinamento do modelo polinomial
    poly_reg = PolynomialFeatures(degree=deg)
    x_train_poly = poly_reg.fit_transform(x_train)
    poly_reg.fit(x_train_poly, y_train)

    lm_poly = LinearRegression()
    lm_poly.fit(x_train_poly, y_train)
    
    #Testando o algoritmo no conjunto de teste
    x_test_poly = poly_reg.fit_transform (x_test)
    y_test_poly_pred = lm_poly.predict(x_test_poly)
    
    #Gráfico da regressão polinomial no conjunto de teste
    plt.figure(figsize=(10,5))
    plt.scatter(x_test, y_test, color = 'r', label='Test set')
    plt.plot(x_test, y_test_poly_pred, color = 'g', label='Prediction')
    plt.xlabel('x_test')
    plt.ylabel('y_test')
    plt.legend()
    plt.title('Regressão Polinomial no conjunto de treinamento ')
    plt.show()
    
    #Calculando e comparando o rmse
    rmse  = np.sqrt(mean_squared_error(y_test, y_test_poly_pred))
    if rmse < min_rmse:
        min_rmse = rmse
        min_deg = deg
```

```python
print('Melhor rmse:', min_rmse, 'obtido com o grau de polinômio = ', min_deg)
```
